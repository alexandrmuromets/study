Найдите наилучшую, с точки зрения среднеквадратичной ошибки, функцию, приближающую зависимость времени разгона автомобиля до 100 км/ч от его массы и мощности двигателя. 


```python
list_symb=[]
for i in range(3):
    list_symb.append(Symbol('a'+str(i)))
list_symb    
    
    
```

```python
list_data=[(1945,560,4.3),(1495,340,4.9),(1570,343,5.2)]
```

```python
sde1=0
for i,k in enumerate(list_data):
    sde2=0
    for j in range(2):
        sde2  = sde2 + k[j]*list_symb[j] 
    sde2  =sde2 + list_symb[2]  
    sde2 = (sde2-k[2])**2
    sde1 = sde1 + sde2

sde = sde1/3    
    
    
```

```python
sde
```

```python
list_dif=[]
for i in list_symb :
    list_dif.append(diff(sde,i))
display(*list_dif)

    
```

```python
nonlinsolve(list_dif,list_symb)
```
